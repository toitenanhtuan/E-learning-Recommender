```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập để hiển thị tất cả các cột
pd.set_option("display.max_columns", None)

# Đọc dữ liệu
file_path = "../data/raw/Coursera.csv"
df = pd.read_csv(file_path)

# 1. Xem thông tin tổng quan
print("----- Thông tin DataFrame -----")
df.info()

# 2. Xem 5 dòng đầu tiên để hiểu cấu trúc
print("\n----- 5 dòng dữ liệu đầu tiên -----")
print(df.head())

# 3. Kiểm tra dữ liệu bị thiếu (missing values)
print("\n----- Số lượng giá trị thiếu mỗi cột -----")
print(df.isnull().sum())

# 4. Phân tích cột 'Difficulty Level'
print("\n----- Phân bố Mức độ Khó -----")
print(df["Difficulty Level"].value_counts())

# 5. Phân tích cột 'Course Rating'
print("\n----- Thống kê Mô tả cho Course Rating -----")
print(df["Course Rating"].describe())

# Quan sát quan trọng nhất: Cột 'Skills' là một chuỗi ký tự, không đồng nhất,
# và có vẻ được phân tách bằng nhiều khoảng trắng.
print("\n----- Ví dụ về dữ liệu cột 'Skills' -----")
print(df["Skills"].iloc[0])
```
